```python
# prompt: load /content/drive/MyDrive/rs-anomic/Modified New/merged_anomaly_data.csv

df_anomaly = pd.read_csv('/content/drive/MyDrive/rs-anomic/Modified New/merged_anomaly_data.csv')
print(df_anomaly.shape)
print("df_anomaly columns:", df_anomaly.columns.tolist())
```

```python
# prompt: load /content/drive/MyDrive/rs-anomic/Modified New/merged_normal_data.csv

df_normal = pd.read_csv('/content/drive/MyDrive/rs-anomic/Modified New/merged_normal_data.csv')
print(df_normal.shape)
print("df_normal columns:", df_normal.columns.tolist())
```

```python
dfs = [df_normal, df_anomaly]

# Concatenate DataFrames vertically
merged_df = pd.concat(dfs, axis=0)

# Reset index if necessary (optional)
merged_df = merged_df.reset_index(drop=True)

print(merged_df.shape)
```

```python
# prompt: drop timestamp column and add id column then give id from 1

# Drop the 'timestamp' column
merged_df = merged_df.drop('timestamp', axis=1)

# Add an 'id' column with values from 1 to the length of the DataFrame
merged_df['id'] = range(1, len(merged_df) + 1)
```

```python
# prompt: print nummber of unique ids

print(merged_df['id'].nunique())
```

```python
# prompt: print number of unique values in each column and their data types as a table

# Calculate unique values and data types for each column
unique_values_info = []
for col in merged_df.columns:
  unique_count = merged_df[col].nunique()
  data_type = merged_df[col].dtype
  unique_values_info.append([col, unique_count, data_type])

# Create a DataFrame for better visualization
unique_values_df = pd.DataFrame(unique_values_info, columns=['Column Name', 'Unique Values', 'Data Type'])

# Display the table
print(unique_values_df)
```

```python
# prompt: drop columns with only one unique value

# Get columns with only one unique value
cols_to_drop = unique_values_df[unique_values_df['Unique Values'] == 1]['Column Name'].tolist()

# Drop these columns from the DataFrame
merged_df = merged_df.drop(columns=cols_to_drop)

print(merged_df.shape)
```

```python
# prompt: plot heatmap for spearman correlation print values too

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate Spearman correlation
corr_matrix = merged_df.corr(method='spearman')

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation Matrix')
plt.show()
```

```python
import numpy as np  # Import numpy

# Find highly correlated features
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# List of features with correlation greater than 0.8
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.8)]

# Drop the highly correlated features
merged_df = merged_df.drop(columns=to_drop)

print(f"Dropped features: {to_drop}")
print(merged_df.shape)
```

```python
# prompt: use Box Plot for outlier detecction except id

# Select all columns except 'id'
columns_to_plot = [col for col in merged_df.columns if col != 'id']

# Create box plots for each column
for col in columns_to_plot:
  plt.figure(figsize=(8, 6))
  sns.boxplot(x=merged_df[col])
  plt.title(f'Box Plot for {col}')
  plt.show()
```

```python

# prompt: use mathematics to fond outliers columns and give

import numpy as np

# Calculate the IQR for each column
def find_outliers_iqr(df):
  outlier_columns = []
  for col in df.select_dtypes(include=np.number).columns:
    if col != 'id':  # Exclude the 'id' column
      Q1 = df[col].quantile(0.25)
      Q3 = df[col].quantile(0.75)
      IQR = Q3 - Q1
      lower_bound = Q1 - 1.5 * IQR
      upper_bound = Q3 + 1.5 * IQR
      outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
      if len(outliers) > 0:
        outlier_columns.append(col)
  return outlier_columns

# Find outlier columns using IQR
outlier_cols = find_outliers_iqr(merged_df)

print("Outlier columns:", outlier_cols)
```

```python
# prompt: if the column has outlier fill with median else fill with mean

# Iterate over the columns
for col in merged_df.columns:
  if col != 'id':  # Exclude the 'id' column
    if col in outlier_cols:
      # Fill outliers with median
      merged_df[col].fillna(merged_df[col].median(), inplace=True)
    else:
      # Fill non-outlier missing values with mean
      merged_df[col].fillna(merged_df[col].mean(), inplace=True)
```

```python
# prompt: use heatmap Plo for missing values

import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of missing values
plt.figure(figsize=(12, 10))
sns.heatmap(merged_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```
